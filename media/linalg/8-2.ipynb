```python
# D. Scott
# Linear Algebra: TCI
# Code Challenge 8.2
# Find null space/left null space basis dimensions and verify 
# summation to ambient dimensions

import random
import numpy as np
from scipy.linalg import null_space

# generate and populate random matrices with same inner value
num1 = []
num2 = []
for i in range(0,144):
    num1.append(random.randint(-6,12))
for i in range(0,99):
    num2.append(random.randint(-6,12))
A1 = np.array(num1).reshape(16,9)
A2 = np.array(num2).reshape(9,11)

A = A1@A2

print("A (16 x 11):\n",A,"\nRank(A) = ",np.linalg.matrix_rank(A))

n = null_space(A)
n_t = null_space(A.transpose())
print("\nNull space basis: \n",n)
print("\nLeft null space basis: \n",n_t)

print("\nNull space dim + row space dim = ",n.shape[1]," + ",np.linalg.matrix_rank(A)," = ",
      n.shape[1]+np.linalg.matrix_rank(A))
print("\nLeft null space dim + column space dim = ",n_t.shape[1]," + ",np.linalg.matrix_rank(A)," = ",
      n_t.shape[1]+np.linalg.matrix_rank(A))
```
