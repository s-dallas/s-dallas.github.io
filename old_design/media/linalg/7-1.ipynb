```python
# D. Scott
# Linear Algebra: TCI
# Code Challenge 7.1
# Generate a random matrix of any particular rank

import random
import numpy as np

# generate and populate random matrices with same inner value
num1 = []
num2 = []
for i in range(0,12):
    num1.append(random.randint(-6,12))
for i in range(0,18):
    num2.append(random.randint(-6,12))
A = np.array(num1).reshape(4,3)
B = np.array(num2).reshape(3,6)

C = A@B

print("A:\n",A)
print("\nB:\n",B)
print("\nC=A@B:\n",C)

# inner value is the rank
print("\nRank(C): ",np.linalg.matrix_rank(C))

# second example
num1 = []
num2 = []
for i in range(0,16):
    num1.append(random.randint(-6,12))
for i in range(0,32):
    num2.append(random.randint(-6,12))
A = np.array(num1).reshape(2,8)
B = np.array(num2).reshape(8,4)

C = A@B

print("A:\n",A)
print("\nB:\n",B)
print("\nC=A@B:\n",C)

print("\nRank(C): ",np.linalg.matrix_rank(C))
```
